### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
ingreso_tienda = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()
```

```python
print(f"Ingreso total Tienda 1: {ingreso_tienda:.2f}")
print(f"Ingreso total Tienda 2: {ingreso_tienda2:.2f}")
print(f"Ingreso total Tienda 3: {ingreso_tienda3:.2f}")
print(f"Ingreso total Tienda 4: {ingreso_tienda4:.2f}")
```

```python
import matplotlib.pyplot as plt

# Prepare data for plotting
tienda_nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_tienda, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(tienda_nombres, ingresos, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])

# Add labels and title
plt.xlabel('Tienda')
plt.ylabel('Ingreso Total')
plt.title('Ingreso Total por Tienda')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Format y-axis labels to be more readable (e.g., in millions or billions)
plt.ticklabel_format(style='plain', axis='y')

# Show the plot
plt.show()
```

# 2. Ventas por categoría

calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
# Ventas por categoría para Tienda 1
ventas_por_categoria_tienda1 = tienda['Categoría del Producto'].value_counts()
print("Categorías más vendidas en Tienda 1:\n", ventas_por_categoria_tienda1.head())

print("\n" + "-"*50 + "\n")

# Ventas por categoría para Tienda 2
ventas_por_categoria_tienda2 = tienda2['Categoría del Producto'].value_counts()
print("Categorías más vendidas en Tienda 2:\n", ventas_por_categoria_tienda2.head())

print("\n" + "-"*50 + "\n")

# Ventas por categoría para Tienda 3
ventas_por_categoria_tienda3 = tienda3['Categoría del Producto'].value_counts()
print("Categorías más vendidas en Tienda 3:\n", ventas_por_categoria_tienda3.head())

print("\n" + "-"*50 + "\n")

# Ventas por categoría para Tienda 4
ventas_por_categoria_tienda4 = tienda4['Categoría del Producto'].value_counts()
print("Categorías más vendidas en Tienda 4:\n", ventas_por_categoria_tienda4.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Combine the sales by category for all stores into a single DataFrame
all_stores_sales = pd.DataFrame({
    'Tienda 1': ventas_por_categoria_tienda1,
    'Tienda 2': ventas_por_categoria_tienda2,
    'Tienda 3': ventas_por_categoria_tienda3,
    'Tienda 4': ventas_por_categoria_tienda4
}).fillna(0) # Fill NaN values with 0 for categories not present in all stores

# Get the top categories overall (e.g., top 8 to cover most common ones)
top_categories_overall = all_stores_sales.sum(axis=1).nlargest(8).index

# Filter the DataFrame to include only these top categories
filtered_sales = all_stores_sales.loc[top_categories_overall]

# Plotting the grouped bar chart
filtered_sales.plot(kind='bar', figsize=(14, 8), width=0.8)
plt.title('Ventas por Categoría en Cada Tienda')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad de Ventas')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.legend(title='Tienda')
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 3. Calificación promedio de la tienda


calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

print(f"Calificación promedio Tienda 1: {calificacion_promedio_tienda1:.2f}")
print(f"Calificación promedio Tienda 2: {calificacion_promedio_tienda2:.2f}")
print(f"Calificación promedio Tienda 3: {calificacion_promedio_tienda3:.2f}")
print(f"Calificación promedio Tienda 4: {calificacion_promedio_tienda4:.2f}")
```

```python
import matplotlib.pyplot as plt

# Prepare data for plotting
calificaciones = [
    calificacion_promedio_tienda1,
    calificacion_promedio_tienda2,
    calificacion_promedio_tienda3,
    calificacion_promedio_tienda4
]

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(calificaciones, labels=tienda_nombres, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.title('Calificación Promedio por Tienda')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

# 4. Productos más y menos vendidos

identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
print("Productos más y menos vendidos por Tienda:\n")

# Tienda 1
ventas_por_producto_tienda1 = tienda['Producto'].value_counts()
print("Top 5 productos más vendidos en Tienda 1:\n", ventas_por_producto_tienda1.head(5))
print("\nTop 5 productos menos vendidos en Tienda 1:\n", ventas_por_producto_tienda1.tail(5))
print("\n" + "-"*50 + "\n")

# Tienda 2
ventas_por_producto_tienda2 = tienda2['Producto'].value_counts()
print("Top 5 productos más vendidos en Tienda 2:\n", ventas_por_producto_tienda2.head(5))
print("\nTop 5 productos menos vendidos en Tienda 2:\n", ventas_por_producto_tienda2.tail(5))
print("\n" + "-"*50 + "\n")

# Tienda 3
ventas_por_producto_tienda3 = tienda3['Producto'].value_counts()
print("Top 5 productos más vendidos en Tienda 3:\n", ventas_por_producto_tienda3.head(5))
print("\nTop 5 productos menos vendidos en Tienda 3:\n", ventas_por_producto_tienda3.tail(5))
print("\n" + "-"*50 + "\n")

# Tienda 4
ventas_por_producto_tienda4 = tienda4['Producto'].value_counts()
print("Top 5 productos más vendidos en Tienda 4:\n", ventas_por_producto_tienda4.head(5))
print("\nTop 5 productos menos vendidos en Tienda 4:\n", ventas_por_producto_tienda4.tail(5))
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Get the single most sold product and its count for each store
most_sold_1 = ventas_por_producto_tienda1.head(1)
most_sold_2 = ventas_por_producto_tienda2.head(1)
most_sold_3 = ventas_por_producto_tienda3.head(1)
most_sold_4 = ventas_por_producto_tienda4.head(1)

# Get the single least sold product and its count for each store
least_sold_1 = ventas_por_producto_tienda1.tail(1)
least_sold_2 = ventas_por_producto_tienda2.tail(1)
least_sold_3 = ventas_por_producto_tienda3.tail(1)
least_sold_4 = ventas_por_producto_tienda4.tail(1)

# Prepare data for plotting most sold products
most_sold_data = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Producto': [most_sold_1.index[0], most_sold_2.index[0], most_sold_3.index[0], most_sold_4.index[0]],
    'Cantidad': [most_sold_1.values[0], most_sold_2.values[0], most_sold_3.values[0], most_sold_4.values[0]]
})

# Prepare data for plotting least sold products
least_sold_data = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Producto': [least_sold_1.index[0], least_sold_2.index[0], least_sold_3.index[0], least_sold_4.index[0]],
    'Cantidad': [least_sold_1.values[0], least_sold_2.values[0], least_sold_3.values[0], least_sold_4.values[0]]
})

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(18, 7))

# Plotting Most Sold Products
axes[0].bar(most_sold_data['Tienda'], most_sold_data['Cantidad'], color='skyblue')
axes[0].set_title('Producto Más Vendido por Tienda')
axes[0].set_xlabel('Tienda')
axes[0].set_ylabel('Cantidad de Ventas')
axes[0].grid(axis='y', linestyle='--', alpha=0.7)
# Add product names on top of bars
for index, row in most_sold_data.iterrows():
    axes[0].text(index, row['Cantidad'], row['Producto'], ha='center', va='bottom', rotation=45)

# Plotting Least Sold Products
axes[1].bar(least_sold_data['Tienda'], least_sold_data['Cantidad'], color='lightcoral')
axes[1].set_title('Producto Menos Vendido por Tienda')
axes[1].set_xlabel('Tienda')
axes[1].set_ylabel('Cantidad de Ventas')
axes[1].grid(axis='y', linestyle='--', alpha=0.7)
# Add product names on top of bars
for index, row in least_sold_data.iterrows():
    axes[1].text(index, row['Cantidad'], row['Producto'], ha='center', va='bottom', rotation=45)

plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

# 5. Envío promedio por tienda

calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python
costo_envio_promedio_tienda1 = tienda['Costo de envío'].mean()
costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

print(f"Costo de envío promedio Tienda 1: {costo_envio_promedio_tienda1:.2f}")
print(f"Costo de envío promedio Tienda 2: {costo_envio_promedio_tienda2:.2f}")
print(f"Costo de envío promedio Tienda 3: {costo_envio_promedio_tienda3:.2f}")
print(f"Costo de envío promedio Tienda 4: {costo_envio_promedio_tienda4:.2f}")
```

```python
import matplotlib.pyplot as plt

# Prepare data for plotting
costos_envio = [
    costo_envio_promedio_tienda1,
    costo_envio_promedio_tienda2,
    costo_envio_promedio_tienda3,
    costo_envio_promedio_tienda4
]

# Create the line chart
plt.figure(figsize=(10, 6))
plt.plot(tienda_nombres, costos_envio, marker='o', linestyle='-', color='purple')

# Add labels and title
plt.xlabel('Tienda')
plt.ylabel('Costo de Envío Promedio')
plt.title('Costo de Envío Promedio por Tienda')
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
